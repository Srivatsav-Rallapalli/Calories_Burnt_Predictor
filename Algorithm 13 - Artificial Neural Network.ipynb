```python
import pandas as pd

import warnings
warnings.simplefilter('ignore')
```

```python
# Importing the Dataset

df = pd.read_csv(r'clean.csv')
df
```

```python
# Seperating the Dependent and Independent variables

X = df.drop(columns=['Calories'])
y = df['Calories']
```

```python
# Splitting the data into Train and Test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=42,train_size=0.8)
```

```python
# Importing the necessary libraries for ANN
import tensorflow as tf
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense

def ann_regressor():
    # Initializing the Network
    model = Sequential()
    # Adding the Input Layer and the First Hidden Layer to the Model
    model.add(Dense(input_dim=3,units=6,activation='relu',kernel_initializer='uniform'))
    # Adding the Second Hidden Layer
    model.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    # Adding the Third Hidden Layer
    model.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    # Adding the Output Layer
    model.add(Dense(units=1,activation='relu',kernel_initializer='uniform'))
    # compiling the Network
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
# Running the Model

ann = ann_regressor()
```

```python
# Train the model using Train data
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```python
# Predicting on Train data

ypred_train = ann.predict(x_train)
ypred_test = ann.predict(x_test)
```

```python
# Calculating the Train r2 score 
from sklearn.metrics import r2_score

print('Train r2 score is :',r2_score(y_train,ypred_train))
print('Test r2 score is :',r2_score(y_test,ypred_test))
```

```python
# Calculating the  cross validation score

def ann_validation():
    regressor = Sequential()
    regressor.add(Dense(input_dim=3,units=6,kernel_initializer='uniform',activation='relu'))
    regressor.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
    regressor.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
    regressor.add(Dense(units=1,kernel_initializer='uniform',activation='relu'))
    regressor.compile(optimizer='adam',loss='mean_squared_error')
    return regressor
```

```python
from scikeras.wrappers import KerasRegressor
regressor = KerasRegressor(ann_validation(),batch_size=32,epochs=100)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressor,x_train,y_train,cv=5)
```
